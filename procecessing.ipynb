```python
import cv2
import matplotlib.pyplot as plt
import os

# ---------------------------
# 1. Load the image (replace with your Kaggle uploaded image path)
# ---------------------------
# In Kaggle, you can upload an image and get the path from /kaggle/input/
image_path = "/kaggle/input/ak-image/gun2.jpg"
image = cv2.imread(image_path)
plt.figure()
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
plt.show()

# ---------------------------
# 2. Convert BGR to RGB
# ---------------------------
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(rgb_image)
plt.title("RGB Image")
plt.axis('off')
plt.show()

# ---------------------------
# 3. Convert RGB image to grayscale
# ---------------------------
gray_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)
plt.figure()
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()

# ---------------------------
# 4. Resize the grayscale image to 300x300 pixels
# ---------------------------
resized_image = cv2.resize(gray_image, (300, 300))
plt.figure()
plt.imshow(resized_image, cmap='gray')
plt.title("Resized Image")
plt.axis('off')
plt.show()

# ---------------------------
# 5. Apply Gaussian blur with a 5x5 kernel
# ---------------------------
blurred_image = cv2.GaussianBlur(resized_image, (5, 5), 0)
plt.figure()
plt.imshow(blurred_image, cmap='gray')
plt.title("Gaussian Blur")
plt.axis('off')
plt.show()

# ---------------------------
# 6. Apply binary thresholding
# ---------------------------
_, threshold_image = cv2.threshold(blurred_image, 127, 255, cv2.THRESH_BINARY)
plt.figure()
plt.imshow(threshold_image, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')
plt.show()

# ---------------------------
# 7. Detect edges using Canny algorithm
# ---------------------------
edges = cv2.Canny(blurred_image, 50, 150)
plt.figure()
plt.imshow(edges, cmap='gray')
plt.title("Canny Edges")
plt.axis('off')
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------
# 1. Load the audio file
# ---------------------------
# Replace with your Kaggle uploaded audio path
audio_path = "/kaggle/input/ak-audio/voice record.mp3"
y, sr = librosa.load(audio_path, sr=None)  # y = audio time series, sr = sample rate
plt.figure()
plt.plot(y)
plt.title("Original Audio Waveform")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()

# ---------------------------
# 2. Convert stereo to mono (if needed)
# ---------------------------
# librosa.load already returns mono by default, but for demonstration:
mono_audio = librosa.to_mono(np.array([y, y])) if len(y.shape) > 1 else y
plt.figure()
plt.plot(mono_audio)
plt.title("Mono Audio Waveform")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()

# ---------------------------
# 3. Normalize audio
# ---------------------------
normalized_audio = mono_audio / np.max(np.abs(mono_audio))
plt.figure()
plt.plot(normalized_audio)
plt.title("Normalized Audio Waveform")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()

# ---------------------------
# 4. Apply a simple low-pass filter (moving average)
# ---------------------------
window_size = 5
smoothed_audio = np.convolve(normalized_audio, np.ones(window_size)/window_size, mode='same')
plt.figure()
plt.plot(smoothed_audio)
plt.title("Smoothed Audio (Low-pass Filter)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()

# ---------------------------
# 5. Compute Short-Time Fourier Transform (STFT)
# ---------------------------
stft = librosa.stft(smoothed_audio)
stft_db = librosa.amplitude_to_db(np.abs(stft))
plt.figure(figsize=(10, 4))
librosa.display.specshow(stft_db, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(format='%+2.0f dB')
plt.title("STFT (Spectrogram)")
plt.show()

# ---------------------------
# 6. Compute Mel-frequency cepstral coefficients (MFCCs)
# ---------------------------
mfccs = librosa.feature.mfcc(y=smoothed_audio, sr=sr, n_mfcc=13)
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
plt.colorbar()
plt.title("MFCCs")
plt.show()

# ---------------------------
# 7. Plot waveform with amplitude envelope
# ---------------------------
frame_length = 2048
hop_length = 512
amplitude_envelope = [max(smoothed_audio[i:i+frame_length]) for i in range(0, len(smoothed_audio), hop_length)]
plt.figure()
plt.plot(smoothed_audio, alpha=0.5)
plt.plot(np.arange(0, len(amplitude_envelope)*hop_length, hop_length), amplitude_envelope, color='r')
plt.title("Waveform with Amplitude Envelope")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```

```python
import re
from nltk.stem import PorterStemmer
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd

# ---------------------------
# Ensure pandas displays full text
# ---------------------------
pd.set_option('display.max_colwidth', None)

# ---------------------------
# 1. Load text
# ---------------------------
text_path = "/kaggle/input/ak-text/ak-text.txt"  # update your path
with open(text_path, 'r', encoding='utf-8') as file:
    text = file.read()

print("Step 1: Original Text Loaded\n")
print(text[:500] + "...\n" if len(text) > 500 else text)
print(f"Total characters: {len(text)}\n")

# ---------------------------
# 2. Convert to lowercase
# ---------------------------
text_lower = text.lower()
print("Step 2: Lowercased Text\n")
print(text_lower[:500] + "...\n" if len(text_lower) > 500 else text_lower)

# ---------------------------
# 3. Remove punctuation and special characters
# ---------------------------
text_clean = re.sub(r'[^a-zA-Z\s]', '', text_lower)
print("Step 3: Cleaned Text (punctuation removed)\n")
print(text_clean[:500] + "...\n" if len(text_clean) > 500 else text_clean)

# ---------------------------
# 4. Tokenize text
# ---------------------------
tokens = text_clean.split()
print("Step 4: Tokenization Completed")
print(f"Total tokens: {len(tokens)}")
print("First 20 tokens:", tokens[:20], "\n")

# ---------------------------
# 5. Remove stopwords (offline list)
# ---------------------------
stop_words = set([
    'i','me','my','myself','we','our','ours','ourselves','you','your','yours',
    'yourself','yourselves','he','him','his','himself','she','her','hers','herself',
    'it','its','itself','they','them','their','theirs','themselves','what','which',
    'who','whom','this','that','these','those','am','is','are','was','were','be','been',
    'being','have','has','had','having','do','does','did','doing','a','an','the','and',
    'but','if','or','because','as','until','while','of','at','by','for','with','about',
    'against','between','into','through','during','before','after','above','below','to',
    'from','up','down','in','out','on','off','over','under','again','further','then',
    'once','here','there','when','where','why','how','all','any','both','each','few',
    'more','most','other','some','such','no','nor','not','only','own','same','so','than',
    'too','very','s','t','can','will','just','don','should','now'
])
tokens_no_stop = [word for word in tokens if word not in stop_words]
print("Step 5: Stopwords Removed")
print(f"Tokens after stopword removal: {len(tokens_no_stop)}")
print("First 20 tokens:", tokens_no_stop[:20], "\n")

# ---------------------------
# 6. Apply stemming
# ---------------------------
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in tokens_no_stop]
print("Step 6: Stemming Applied")
print("First 20 stemmed tokens:", stemmed_tokens[:20], "\n")

# ---------------------------
# 7. Compute word frequencies
# ---------------------------
word_freq = Counter(stemmed_tokens)
print("Step 7: Word Frequency Computed")
print("Top 10 words:", word_freq.most_common(10), "\n")

# ---------------------------
# 8. Display top 10 words as bar chart
# ---------------------------
top_words = word_freq.most_common(10)
words = [w[0] for w in top_words]
counts = [w[1] for w in top_words]

plt.figure(figsize=(8,5))
plt.bar(words, counts, color='skyblue')
plt.title("Top 10 Words Frequency")
plt.xlabel("Words")
plt.ylabel("Counts")
plt.show()

# ---------------------------
# 9. Display all steps in a table
# ---------------------------
df_steps = pd.DataFrame({
    "Step": [
        "Original Text",
        "Lowercased Text",
        "Cleaned Text",
        "Tokens",
        "Tokens without Stopwords",
        "Stemmed Tokens"
    ],
    "Output": [
        text,
        text_lower,
        text_clean,
        ' '.join(tokens),
        ' '.join(tokens_no_stop),
        ' '.join(stemmed_tokens)
    ],
    "Count": [
        len(text),
        len(text_lower),
        len(text_clean),
        len(tokens),
        len(tokens_no_stop),
        len(stemmed_tokens)
    ]
})

print("Step 9: All processing steps summary\n")
df_steps

# ---------------------------
# 10. Display Original vs Processed Text side by side
# ---------------------------
df_compare = pd.DataFrame({
    "Original Text": [text],
    "Processed Text": [' '.join(stemmed_tokens)]
})

print("Step 10: Original vs Processed Text Comparison\n")
df_compare
```
